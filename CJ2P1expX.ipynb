```python
#import libraries
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#this function returns the exponential function
def myFunction(x):
    y = np.exp(x)
    return(y)
```

```python
#this is the main function
def main():
    x = np.linspace(0,1,100)
    fig=plt.figure()
    plt.plot(x,myFunction(x))
    plt.xlabel('Time(Milliseconds)')
    plt.ylabel('Awesomeness')
    plt.savefig('CJ2P1expX.pdf',bbox_inches='tight',dpi=400)
```

```python
#execute main
if __name__ =="__main__":
    main()
```
