```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = 1000
x = np.random.triangular(left=0.0,right=1.0,mode=0.5,size=y)

#create the histograms
plt.hist(x,bins=100,edgecolor="black")
plt.ylabel("N per bin")
plt.xlabel("x")
plt.savefig("CJ2P4Hist_other.pdf",bbox_inches="tight",facecolor="white")
```
