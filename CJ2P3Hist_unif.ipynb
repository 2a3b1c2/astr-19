```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = 1000
x = np.random.uniform(low=0.0,high=1.0,size=y)

#create the histograms

plt.hist(x,bins=100,edgecolor="black")
plt.ylabel("N per bin")
plt.xlabel("x")
plt.savefig("CJ2P3Hist_unif.pdf",bbox_inches="tight",facecolor="white")
```
